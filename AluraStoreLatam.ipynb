# 1. Sección 1: Manipulación y Limpieza de Datos

###1- Importación de datos



```python
import pandas as pd # Importa la librería pandas, comúnmente utilizada para manipulación y análisis de datos.

# Define las URLs de los archivos CSV para cada tienda.
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Lee los archivos CSV desde las URLs y los carga en dataframes de pandas.
tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

```

###2.4. Facturación total por categoría

```python
# 2.4 Cálculo de facturación total por categoría

# Agrupamos por categoría y sumamos los precios de venta
facturacion_categoria = (
    df.groupby('Categoría del Producto', as_index=False)['Precio']
      .sum()
      .rename(columns={'Precio': 'Facturación'})
      .sort_values('Facturación', ascending=False)
)

# Mostramos la tabla resultante
print("Facturación total por categoría:\n")
print(facturacion_categoria)
```

```python
# 2.4.1 Mostrar la facturación total por categoría con valores reales

# Creamos una copia para formatear sin alterar los datos numéricos
facturacion_categoria_fmt = facturacion_categoria.copy()

# Mostramos los valores con formato monetario (sin cambiar su escala real)
print("Facturación total por categoría (valores reales en $):\n")
for i, row in facturacion_categoria_fmt.iterrows():
    print(f"{row['Categoría del Producto']:<30} ${row['Facturación']:,.0f}")

```

```python
# 2.4.2 Combinar cantidad de registros y facturación total por categoría

# Calculamos el conteo de registros por categoría
conteo_categorias = df['Categoría del Producto'].value_counts().reset_index()
conteo_categorias.columns = ['Categoría del Producto', 'Cantidad de Ventas']

# Calculamos la facturación total por categoría
facturacion_categoria = (
    df.groupby('Categoría del Producto', as_index=False)['Precio']
      .sum()
      .rename(columns={'Precio': 'Facturación'})
)

# Unimos ambos resultados por la columna 'Categoría del Producto'
resumen_categoria = pd.merge(conteo_categorias, facturacion_categoria, on='Categoría del Producto')

# Ordenamos de mayor a menor facturación
resumen_categoria = resumen_categoria.sort_values('Facturación', ascending=False)

# Mostramos la tabla formateada
print("Resumen de ventas y facturación por categoría:\n")
for i, row in resumen_categoria.iterrows():
    print(f"{row['Categoría del Producto']:<30} {row['Cantidad de Ventas']:>6}  ${row['Facturación']:>15,.0f}")
```

###2.5 Visualización de Ventas por Categoría


#### 2.5.1. Gráfico de facturación total por categoría

```python
# 2.5.1 Gráfico de facturación total por categoría (en millones de $)
import matplotlib.pyplot as plt

# Copiamos y convertimos la facturación a millones
fact_cat_plot = facturacion_categoria.copy()
fact_cat_plot['Facturación (M)'] = fact_cat_plot['Facturación'] / 1e6

# Ordenamos de menor a mayor para el gráfico horizontal
fact_cat_plot = fact_cat_plot.sort_values('Facturación (M)', ascending=True)

# Configuración del gráfico
plt.figure(figsize=(10,6))
barras = plt.barh(
    fact_cat_plot['Categoría del Producto'],
    fact_cat_plot['Facturación (M)'],
    color='#1f77b4',
    edgecolor='black'
)

# Etiquetas con valores en millones
for barra in barras:
    valor = barra.get_width()
    plt.text(valor + 50,  # pequeño desplazamiento a la derecha
             barra.get_y() + barra.get_height()/2,
             f"${valor:,.1f}M",
             va='center', fontsize=9, fontweight='bold', color='#333333')

# Títulos y etiquetas
plt.title("Facturación total por categoría", fontsize=14, fontweight='bold')
plt.xlabel("Facturación (en millones de $)")
plt.ylabel("Categorías")

plt.tight_layout()
plt.show()
```

```python
# 2.5.2-A Ventas en unidades (barras horizontales)
import matplotlib.pyplot as plt

# Ordenamos de menor a mayor para que la barra más alta quede arriba
ventas_plot = resumen_categoria.sort_values('Cantidad de Ventas', ascending=True).copy()

plt.figure(figsize=(10,6))
barras = plt.barh(
    ventas_plot['Categoría del Producto'],
    ventas_plot['Cantidad de Ventas'],
    color='#156082', edgecolor='black'
)

# Etiquetas con valores al final de cada barra
for b in barras:
    w = b.get_width()
    plt.text(w + ventas_plot['Cantidad de Ventas'].max() * 0.02,
             b.get_y() + b.get_height()/2,
             f"{w:,.0f}", va='center', fontsize=9, fontweight='bold')

# Títulos y etiquetas
plt.title("Ventas por categoría (en unidades)", fontsize=14, fontweight='bold')
plt.xlabel("Cantidad de ventas (unidades)")
plt.ylabel("Categorías")

plt.tight_layout()
plt.show()
```

```python
# 2.5.2-B Facturación por categoría (barras horizontales, en millones de $)
import matplotlib.pyplot as plt

# Convertimos a millones y ordenamos (desc)
fact_plot = resumen_categoria.assign(**{
    'Facturación (M)': resumen_categoria['Facturación'] / 1e6
}).sort_values('Facturación (M)', ascending=True)  # asc para que el horizontal muestre mayor arriba

plt.figure(figsize=(10,6))
barras = plt.barh(
    fact_plot['Categoría del Producto'],
    fact_plot['Facturación (M)'],
    color='#1f77b4', edgecolor='black'
)

# Etiquetas al final de cada barra
for b in barras:
    w = b.get_width()
    plt.text(w + fact_plot['Facturación (M)'].max()*0.02,
             b.get_y() + b.get_height()/2,
             f"${w:,.1f}M", va='center', fontsize=9, fontweight='bold')

plt.title("Facturación por categoría (millones de $)", fontsize=14, fontweight='bold')
plt.xlabel("Facturación (millones de $)")
plt.ylabel("Categorías")
plt.tight_layout()
plt.show()
```

```python
# 2.5.3 Participación porcentual de la facturación por categoría (dona)
import matplotlib.pyplot as plt
import numpy as np

# --- 1) Base: usamos la tabla 'resumen_categoria' (tiene 'Categoría del Producto' y 'Facturación')
part = resumen_categoria[['Categoría del Producto','Facturación']].copy()

# --- 2) Ordenamos de mayor a menor facturación (mejor lectura en leyenda)
part = part.sort_values('Facturación', ascending=False).reset_index(drop=True)

# --- 3) Cálculo del total (para anotarlo) y del porcentaje
total_fact = part['Facturación'].sum()
pct = part['Facturación'] / total_fact * 100

# --- 4) Colores (paleta consistente con los gráficos anteriores)
colors = ['#1f77b4','#156082','#ff7f0e','#2ca02c','#d62728','#9467bd','#8c564b','#e377c2']

# --- 5) Gráfico de dona
fig, ax = plt.subplots(figsize=(8,6))

# Etiquetas dentro del gráfico: solo porcentaje (limpias)
def autopct_pct(p):
    return f"{p:.1f}%"

wedges, texts, autotexts = ax.pie(
    part['Facturación'],
    labels=None,                   # etiquetas en leyenda, no sobre el slice
    autopct=autopct_pct,
    startangle=90,
    colors=colors[:len(part)],
    pctdistance=0.78,
    wedgeprops=dict(edgecolor='white', linewidth=1)
)

# Efecto dona (agujero central)
centre_circle = plt.Circle((0,0), 0.58, fc='white')
fig.gca().add_artist(centre_circle)

# --- 6) Leyenda a la derecha con categoría + % y facturación en M$
labels_legend = [
    f"{cat} — {v/1e6:,.1f}M  ({p:.1f}%)"
    for cat, v, p in zip(part['Categoría del Producto'], part['Facturación'], pct)
]
ax.legend(wedges, labels_legend, title="Categorías", loc="center left", bbox_to_anchor=(1.02, 0.5), frameon=False)

# --- 7) Título y anotación del total en el centro
ax.set_title("Participación de la facturación por categoría", fontsize=14, fontweight='bold')
ax.text(0, 0, f"Total\n${total_fact/1e6:,.1f}M", ha='center', va='center', fontsize=11, fontweight='bold', color='#333333')

plt.tight_layout()
plt.show()
```

###2.6. Ventas por categoría y tienda

####2.6.1 — Tabla: cantidad de ventas por categoría y tienda

```python
# 2.6.1 Ventas por categoría y tienda
ventas_cat_tienda = (
    df.groupby(['Tienda', 'Categoría del Producto'], as_index=False)
      .size()
      .rename(columns={'size': 'Cantidad de Ventas'})
      .sort_values(['Tienda', 'Cantidad de Ventas'], ascending=[True, False])
)

# Mostramos todas las filas (no solo las primeras)
pd.set_option('display.max_rows', None)

print("Cantidad de productos vendidos por categoría y tienda (valores reales):\n")
print(ventas_cat_tienda)
```

####2.6.2 — Facturación total por categoría y tienda

```python
# 2.6.2 Facturación total y cantidad de ventas por categoría y tienda
fact_cat_tienda = (
    df.groupby(['Tienda', 'Categoría del Producto'], as_index=False)
      .agg(
          Cantidad_de_Ventas=('Precio', 'count'),
          Facturacion_Total=('Precio', 'sum')
      )
      .sort_values(['Tienda', 'Facturacion_Total'], ascending=[True, False])
)

# Formateamos facturación con separador de miles
fact_cat_tienda['Facturacion_Total'] = fact_cat_tienda['Facturacion_Total'].apply(lambda x: f"${x:,.0f}")

# Mostramos todas las filas
pd.set_option('display.max_rows', None)
print("Facturación total y cantidad de ventas por categoría y tienda (valores reales):\n")
print(fact_cat_tienda)
```

####2.6.3-A — Heatmap de cantidad de ventas (Tienda × Categoría)

```python
# 2.6.3-A Heatmap de CANTIDAD de ventas por Tienda × Categoría
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# 1) Matriz de cantidades
mat_cnt = (
    df.groupby(['Tienda','Categoría del Producto'])
      .size()
      .unstack(fill_value=0)
)

# Ordenar categorías por total global (desc) para lectura consistente
orden_cats = mat_cnt.sum(axis=0).sort_values(ascending=False).index
mat_cnt = mat_cnt[orden_cats]

# 2) Gráfico
fig, ax = plt.subplots(figsize=(10, 4.8))
im = ax.imshow(mat_cnt.values, cmap='Blues')

# Ejes y etiquetas
ax.set_xticks(np.arange(mat_cnt.shape[1]))
ax.set_yticks(np.arange(mat_cnt.shape[0]))
ax.set_xticklabels(mat_cnt.columns, rotation=25, ha='right', fontsize=8)
ax.set_yticklabels(mat_cnt.index, fontsize=10)

# Anotaciones con valores
for i in range(mat_cnt.shape[0]):
    for j in range(mat_cnt.shape[1]):
        ax.text(j, i, f"{mat_cnt.iloc[i, j]:,}", ha='center', va='center', color='#123', fontsize=8, fontweight='bold')

# Título y colorbar
ax.set_title("Cantidad de ventas por categoría y tienda", fontsize=14, fontweight='bold')
cbar = fig.colorbar(im, ax=ax)
cbar.set_label("Ventas (unidades)", rotation=90)

plt.tight_layout()
plt.show()
```

###2.6.3-B — Heatmap de facturación (en millones de $) (Tienda × Categoría)

```python
# 2.6.3-B Heatmap de FACTURACIÓN por Tienda × Categoría (en millones de $)
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Asegurar columna numérica cruda (como en la auditoría)
if '_precio_num' not in df.columns:
    df['_precio_num'] = df['Precio'].astype(float)

# 1) Matriz de facturación en millones
mat_val = (
    df.groupby(['Tienda','Categoría del Producto'])['_precio_num']
      .sum()
      .unstack(fill_value=0) / 1e6
)

# Mismo orden de categorías que en el heatmap de unidades
mat_val = mat_val[orden_cats]

# 2) Gráfico
fig, ax = plt.subplots(figsize=(10, 4.8))
im = ax.imshow(mat_val.values, cmap='Oranges')

# Ejes y etiquetas
ax.set_xticks(np.arange(mat_val.shape[1]))
ax.set_yticks(np.arange(mat_val.shape[0]))
ax.set_xticklabels(mat_val.columns, rotation=25, ha='right', fontsize=8)
ax.set_yticklabels(mat_val.index, fontsize=10)

# Anotaciones con valores (M$)
for i in range(mat_val.shape[0]):
    for j in range(mat_val.shape[1]):
        ax.text(j, i, f"${mat_val.iloc[i, j]:,.1f}M", ha='center', va='center', color='#4a2a00', fontsize=8, fontweight='bold')

# Título y colorbar
ax.set_title("Facturación por categoría y tienda (millones de $)", fontsize=14, fontweight='bold')
cbar = fig.colorbar(im, ax=ax)
cbar.set_label("Facturación (M$)")

plt.tight_layout()
plt.show()
```

####2.6.3-D — Top 3 categorías por tienda (por facturación total real)

```python
# 2.6.3-D Top 3 categorías por TIENDA (por facturación real)
import pandas as pd

# Asegurar columna numérica cruda (por si no está creada)
if '_precio_num' not in df.columns:
    df['_precio_num'] = df['Precio'].astype(float)

# Agrupamos y sumamos la facturación por tienda y categoría
top3_fact = (
    df.groupby(['Tienda', 'Categoría del Producto'])['_precio_num']
      .sum()
      .reset_index(name='Facturación')
      .sort_values(['Tienda', 'Facturación'], ascending=[True, False])
      .groupby('Tienda')
      .head(3)
)

# Mostramos el resultado
print("Top 3 categorías por tienda (por facturación total):\n")
for tienda, sub in top3_fact.groupby('Tienda'):
    print(f"\n{tienda}")
    for _, r in sub.iterrows():
        print(f"   - {r['Categoría del Producto']}: ${r['Facturación']:,.0f}")
```

####3.1 – Calificación promedio por tienda

```python
# 3.1 Calificación promedio por tienda (orden lógico de tiendas)
import pandas as pd
from pandas.api.types import CategoricalDtype

# Definir orden lógico de tiendas
orden_tiendas = CategoricalDtype(['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'], ordered=True)

# Calcular métricas
calif_tienda = (
    df.groupby('Tienda', as_index=False)['Calificación']
      .agg(Promedio='mean', Mínimo='min', Máximo='max', Total_Registros='count')
)

# Aplicar orden y redondeo
calif_tienda['Tienda'] = calif_tienda['Tienda'].astype(orden_tiendas)
calif_tienda = calif_tienda.sort_values('Tienda')
calif_tienda['Promedio'] = calif_tienda['Promedio'].round(2)


print("Calificación promedio de los clientes por tienda (orden lógico):\n")
print(calif_tienda)

```

####3.2 – Métricas de satisfacción detalladas)

```python
# 3.1.2 Métricas de satisfacción detalladas por tienda
import pandas as pd
import numpy as np

# Asegurar columna numérica
df['Calificación'] = df['Calificación'].astype(int)

# Calcular métricas
calif_detallada = (
    df.groupby('Tienda')['Calificación']
      .agg(
          Promedio='mean',
          Desviación='std',
          Mínimo='min',
          Máximo='max',
          Moda=lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan,
          Total='count',
          Satisfechos=lambda x: (x >= 4).mean() * 100,
          Muy_Satisfechos=lambda x: (x == 5).mean() * 100
      )
      .round(2)
      .reset_index()
)

print("Indicadores detallados de satisfacción por tienda (datos reales):\n")
print(calif_detallada)

```

####3.3 – Gráfico comparativo de calificaciones promedio por tienda

```python
import matplotlib.pyplot as plt
import pandas as pd

# Datos reales de calificación promedio
data = calif_detallada[['Tienda', 'Promedio']].copy()
data = data.sort_values('Promedio', ascending=False)

# Colores: gris azulado para todas, rojo para la peor
colores = ['#7EA8BE' if val > data['Promedio'].min() else 'red' for val in data['Promedio']]

# Gráfico de barras
plt.figure(figsize=(8,5))
bars = plt.bar(data['Tienda'], data['Promedio'], color=colores, edgecolor='black')

# Etiquetas de valores arriba de cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.01, f"{yval:.2f}", ha='center', va='bottom', fontsize=11, fontweight='bold')

# Título y estilo
plt.title("CALIFICACIÓN PROMEDIO POR TIENDA", fontsize=14, fontweight='bold')
plt.ylim(3.8, 4.1)
plt.ylabel("Calificación promedio")
plt.xlabel("")
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# 1️⃣ Agrupar y ordenar los productos por tienda según cantidad de ventas
ventas_producto_ordenado = (
    df.groupby(['Tienda', 'Producto'])
      .size()
      .reset_index(name='Cantidad_Ventas')
      .sort_values(['Tienda', 'Cantidad_Ventas'], ascending=[True, False])
)

# 2️⃣ Obtener el más vendido por tienda
top_productos = (
    ventas_producto_ordenado.groupby('Tienda')
    .head(1)
    .reset_index(drop=True)
)

# 3️⃣ Obtener el menos vendido por tienda
bottom_productos = (
    ventas_producto_ordenado.groupby('Tienda')
    .tail(1)
    .reset_index(drop=True)
)

# 4️⃣ Combinar ambos resultados en una sola tabla
comparativo = pd.merge(
    top_productos,
    bottom_productos,
    on='Tienda',
    suffixes=('_MasVendido', '_MenosVendido')
)

# 5️⃣ Mostrar resultados
print("📊 Productos más y menos vendidos por tienda:\n")
for _, row in comparativo.iterrows():
    print(f"{row['Tienda']}:")
    print(f"   🟢 Más vendido  → {row['Producto_MasVendido']} ({row['Cantidad_Ventas_MasVendido']} unidades)")
    print(f"   🔴 Menos vendido → {row['Producto_MenosVendido']} ({row['Cantidad_Ventas_MenosVendido']} unidades)\n")
```

####7. Análisis Financiero Comparativo de Tiendas

####7.1 Identificar que tienda genera mayor o menor rentabilidad neta

```python
# --- Análisis financiero comparativo por tienda  ---

finanzas = (
    df.groupby('Tienda', as_index=False)
      .agg({
          'Precio': 'sum',
          'Costo de envío': 'sum'
      })
      .rename(columns={'Precio': 'Ingresos_Totales', 'Costo de envío': 'Costo_Total_Envio'})
)

# Cálculo del margen operativo y porcentaje
finanzas['Margen_Operativo'] = finanzas['Ingresos_Totales'] - finanzas['Costo_Total_Envio']
finanzas['Margen_%'] = (finanzas['Margen_Operativo'] / finanzas['Ingresos_Totales']) * 100

# Formateo numérico (sin notación científica)
finanzas_fmt = finanzas.copy()
for col in ['Ingresos_Totales', 'Costo_Total_Envio', 'Margen_Operativo']:
    finanzas_fmt[col] = finanzas_fmt[col].apply(lambda x: f"${x:,.0f}")

finanzas_fmt['Margen_%'] = finanzas['Margen_%'].round(2)

# Mostrar tabla ordenada
finanzas_fmt = finanzas_fmt.sort_values('Margen_%', ascending=False).reset_index(drop=True)

print("Análisis financiero comparativo por tienda (valores formateados):\n")
print(finanzas_fmt.to_string(index=False))
```

####Ticket Promedio

```python
# --- Métricas de rentabilidad por tienda ---

# Cálculo base
ventas_por_tienda = df.groupby('Tienda', as_index=False).agg({
    'Precio': 'sum',
    'Costo de envío': 'sum',
    'Producto': 'count'  # número total de transacciones o ventas
}).rename(columns={'Precio': 'Ingresos_Totales',
                   'Costo de envío': 'Costo_Total_Envio',
                   'Producto': 'Total_Ventas'})

# Nuevas métricas
ventas_por_tienda['Ticket_Promedio'] = ventas_por_tienda['Ingresos_Totales'] / ventas_por_tienda['Total_Ventas']
ventas_por_tienda['Costo_Promedio_Envio'] = ventas_por_tienda['Costo_Total_Envio'] / ventas_por_tienda['Total_Ventas']
ventas_por_tienda['Margen_Promedio_Venta'] = ventas_por_tienda['Ticket_Promedio'] - ventas_por_tienda['Costo_Promedio_Envio']

# Formato numérico
ventas_fmt = ventas_por_tienda.copy()
for col in ['Ingresos_Totales', 'Costo_Total_Envio', 'Ticket_Promedio', 'Costo_Promedio_Envio', 'Margen_Promedio_Venta']:
    ventas_fmt[col] = ventas_fmt[col].apply(lambda x: f"${x:,.0f}")

# Ordenar por margen promedio
ventas_fmt = ventas_fmt.sort_values('Margen_Promedio_Venta', ascending=False).reset_index(drop=True)

print("Indicadores financieros por tienda (valores reales):\n")
print(ventas_fmt.to_string(index=False))
```

####Análisis de Participación Relativa ("Peso sobre el Total")

```python
# --- Análisis de peso relativo sobre el total del negocio ---

# Base de cálculo: métricas principales por tienda
peso_tiendas = (
    df.groupby('Tienda', as_index=False)
      .agg({
          'Precio': 'sum',
          'Costo de envío': 'sum',
          'Producto': 'count'
      })
      .rename(columns={'Precio': 'Ingresos_Totales',
                       'Costo de envío': 'Costo_Total_Envio',
                       'Producto': 'Total_Ventas'})
)

# Cálculo de márgenes
peso_tiendas['Margen_Operativo'] = peso_tiendas['Ingresos_Totales'] - peso_tiendas['Costo_Total_Envio']

# Totales globales (para porcentajes)
tot_ingresos = peso_tiendas['Ingresos_Totales'].sum()
tot_costos = peso_tiendas['Costo_Total_Envio'].sum()
tot_margen = peso_tiendas['Margen_Operativo'].sum()
tot_ventas = peso_tiendas['Total_Ventas'].sum()

# Cálculo de pesos porcentuales
peso_tiendas['% Ingresos'] = (peso_tiendas['Ingresos_Totales'] / tot_ingresos * 100).round(2)
peso_tiendas['% Costos'] = (peso_tiendas['Costo_Total_Envio'] / tot_costos * 100).round(2)
peso_tiendas['% Margen'] = (peso_tiendas['Margen_Operativo'] / tot_margen * 100).round(2)
peso_tiendas['% Ventas'] = (peso_tiendas['Total_Ventas'] / tot_ventas * 100).round(2)

# Formato monetario para legibilidad
for col in ['Ingresos_Totales', 'Costo_Total_Envio', 'Margen_Operativo']:
    peso_tiendas[col] = peso_tiendas[col].apply(lambda x: f"${x:,.0f}")

# Mostrar tabla ordenada por % de ingresos
peso_tiendas = peso_tiendas.sort_values('% Ingresos', ascending=False).reset_index(drop=True)

print("Distribución porcentual del desempeño por tienda:\n")
print(peso_tiendas.to_string(index=False))
```

####Análisis de Participación Relativa ("Peso sobre el Total")
